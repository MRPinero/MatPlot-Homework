```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = r"C:\Users\melis\Desktop\MatPlot-Homework\Pyber\city_data.csv"
ride_data_to_load = r"C:\Users\melis\Desktop\MatPlot-Homework\Pyber\ride_data.csv"

#Read the City and Ride Data
city_data_df=pd.read_csv(city_data_to_load)
city_data_df.head()

ride_data_df=pd.read_csv(ride_data_to_load)
ride_data_df.head()

# Combine the data into a single dataset
pyber_data=pd.merge(city_data_df, ride_data_df, on="city")
pyber_data.head()
# Display the data table for preview
```

```python
#Average Fare ($) Per City
average_fare=pyber_data.groupby("city")["fare"].mean()
average_fare
```

```python
#Total Number of Rides per City
total_rides=pyber_data.groupby("city")["ride_id"].count()
total_rides
```

```python
#Total Number of Drivers per City
total_drivers=pyber_data.groupby("city")["driver_count"].mean()
total_drivers
```

```python
#City Type (Urban, Suburban, Rural)
city_type=city_data_df.set_index("city")["type"]
city_type.value_counts()
```

```python
pyber_cities=pd.DataFrame({"Average Fare Per City": average_fare, "Number of Rides": total_rides, "Number of Drivers": total_drivers, "City Type": city_type})
pyber_cities.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types


# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
# Show plot
urban=pyber_cities.loc[pyber_cities["City Type"]== "Urban"]
suburban=pyber_cities.loc[pyber_cities["City Type"]== "Suburban"]
rural=pyber_cities.loc[pyber_cities["City Type"]== "Rural"]

plt.scatter(urban["Number of Rides"], urban["Average Fare Per City"], color="gold", edgecolors="black", s=urban["Number of Drivers"]*16, label="Urban", alpha=0.5, linewidth=0.5)
plt.show()

plt.scatter(suburban["Number of Rides"], suburban["Average Fare Per City"], color="skyblue", edgecolors="black", s=suburban["Number of Drivers"]*16, label="Urban", alpha=0.5, linewidth=0.5)
plt.show()

plt.scatter(rural["Number of Rides"], rural["Average Fare Per City"], color="coral", edgecolors="black", s=rural["Number of Drivers"]*16, label="Urban", alpha=0.5, linewidth=1.5)
plt.show()

#plot title
plt.title("Average Fare/City vs Number of Rides")

#axes labels
plt.xlabel("Number of Rides")
plt.ylabel("Average Fare Per City")
label=["Urban", "Suburban", "Rural"]

#legend labels
plt.legend(label, loc="upper right", title="City Type", ncol=1, fontsize=8)

#Show Bubble Plot
plt.scatter(urban["Number of Rides"], urban["Average Fare Per City"], color="gold", edgecolors="black", s=urban["Number of Drivers"]*16, label="Urban", alpha=0.5, linewidth=0.5)
plt.scatter(suburban["Number of Rides"], suburban["Average Fare Per City"], color="skyblue", edgecolors="black", s=suburban["Number of Drivers"]*16, label="Urban", alpha=0.5, linewidth=0.5)
plt.scatter(rural["Number of Rides"], rural["Average Fare Per City"], color="coral", edgecolors="black", s=rural["Number of Drivers"]*16, label="Urban", alpha=0.5, linewidth=1.5)
plt.show


```
